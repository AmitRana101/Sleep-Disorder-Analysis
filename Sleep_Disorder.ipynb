# Preprocessing

```python
df1['Occupation'].value_counts().plot(kind='bar')
plt.title("Occupation count")
plt.show()
```

```python
# use map for calculate in blood pressure
df1[["Systolic","Diastolic"]] = df1['Blood Pressure'].str.split('/',expand=True).astype(int)
df1.head(1)
```

```python
# now calculat map value(its range of Blood Pressure)
# MAP = DP + 1/3(SP – DP)
df1["Map"]= round(df1["Diastolic"] + 1/3*(df1["Systolic"] - df1["Diastolic"]),1)
df1.head(1)
```

```python
df1['Map'].astype(int)
```

```python
# which one is better in good sleep and physical healt and heart rate,
# filter healthy indiindividual:
a=round(df1['Quality of Sleep'].mean(),2)
b=round(df1['Physical Activity Level'].mean(),2)
c=round(df1['Heart Rate'].mean(),2)
print(a,b,c)

# 
Healthy_occu = (df1['Quality of Sleep']>=a) & \
               (df1['Physical Activity Level']>=b) & \
               (df1['Heart Rate']<=c)
Health_valu = df1[Healthy_occu]['Occupation'].value_counts()
print(f"Their has healthy person with job {Health_valu}")
```

```python
col = df1['Occupation'].value_counts()
plt.pie(Health_valu,labels=df1[Healthy_occu]['Occupation'].value_counts(),autopct = '%.2f%%',explode=[.1,.1,.1,.1,.1,.1,.1],shadow=True )
plt.show()
```

```python
# how many doctor suffering for isomina disorder and Systolic.
insomina_doc = df1[(df1['Occupation']=='Doctor') & (df1['Sleep Disorder']=='Insomnia')].shape[0]

high_blood_pressure = df1[(df1['Occupation']=='Doctor') & (df1['Systolic'])].shape[0]
print(f"The doctor suffering insomina {insomina_doc}")
print(f"The doctor suffering high blood pressure {high_blood_pressure}")
```

```python
# how many gender sufferig of high bp and low bp.
result = df1.groupby('Gender').apply(lambda x:pd.Series({
    "high_bp" : (x['Systolic']>=x['Map']).sum(),
    "low_bp" : (x['Diastolic']<x['Map']).sum(),
    "total_bp" : ((x['Systolic']>x['Map']) | (x['Diastolic']>x['Map'])).sum()}))
print(result)
```
